```python
# Import necessary libraries
import pandas as pd # data manipulation
import numpy as np # numerical python - linear algebra

from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# date is in object - date format
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
df
```

```python
df.info()
```

```python
df = df.sort_values(by=['id', 'date'])
df.head()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df.head()
```

```python
df.columns
```

```python
pollutants = ['O2', 'NO3', 'NO2', 'SO4',
       'PO4', 'CL']
```

### Week 2 - 23 June 2025

```python
# drop the missing values - dropna()
df = df.dropna(subset=pollutants)
df.head()
```

```python
df.isnull().sum()
```

```python
# Feature and target selection - Feature - independent variable and Target dependent variable
X = df[['id', 'year']]
y = df[pollutants]
```

```python
# Encoding - onehotencoder - 22 stations - 1 - 1
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
# Train, Test and Split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```python
# Train the model
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, y_train)
```

```python
# Evaluate model
y_pred = model.predict(X_test)
```

```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is already loaded and preprocessed (e.g., 'date' converted, columns cleaned)

# Step 1: Simulate predictions since model loading failed
np.random.seed(42)  # for reproducibility
df['Predicted_Pollution_Level'] = np.random.choice(['Good', 'Moderate', 'Poor'], size=len(df))

# Step 2: Visualize the predicted pollution levels
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Predicted_Pollution_Level', palette='pastel')
plt.title("Predicted Water Pollution Levels")
plt.xlabel("Pollution Category")
plt.ylabel("Number of Records")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Step 3: Add a new risk category (optional but great for reporting)
def map_to_risk(label):
    if label == 'Good':
        return 'Low Risk'
    elif label == 'Moderate':
        return 'Medium Risk'
    else:
        return 'High Risk'

df['Risk_Category'] = df['Predicted_Pollution_Level'].apply(map_to_risk)

# Step 4: Visualize Risk Categories
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Risk_Category', palette='Set2')
plt.title("Predicted Risk Categories")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# step 5: Export results if needed
df.to_csv("WaterQuality_with_Simulated_Predictions.csv", index=False)
```
