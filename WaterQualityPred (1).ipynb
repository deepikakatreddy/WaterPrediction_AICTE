```python
#distribution of NH4
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['NH4'], kde=True, color='skyblue')
plt.title("Distribution of NH4 (Ammonium Levels)")
plt.xlabel("NH4 (mg/L)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract 'year' from 'date'
df['year'] = df['date'].dt.year

# Line plot: Average BSK5 per year
plt.figure(figsize=(10, 6))
df.groupby('year')['BSK5'].mean().plot(kind='line', marker='o', color='green')
plt.title("Average BSK5 Over the Years")
plt.xlabel("Year")
plt.ylabel("BSK5 (mg/L)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Convert 'date' to datetime format
df['date'] = pd.to_datetime(df['date'])

# Step 2: Extract 'month' from 'date'
df['month'] = df['date'].dt.month

# Step 3: Plot Monthly Trend of Suspended Solids
plt.figure(figsize=(10, 6))
df.groupby('month')['Suspended'].mean().plot(kind='bar', color='orange')
plt.title("Average Suspended Solids by Month")
plt.xlabel("Month")
plt.ylabel("Suspended Solids (mg/L)")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
